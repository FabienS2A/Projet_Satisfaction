```python
import pandas as pd
import numpy as np

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score  

train_data = pd.read_csv("Train_Data.csv")
test_data = pd.read_csv("Test_Data.csv")
X_train = train_data.drop("Satisfaction", axis=1)
X_test = test_data.drop("Satisfaction", axis=1)
y_train = train_data["Satisfaction"]
y_test = test_data["Satisfaction"]

                            # Régression linéaire
# model_LiR = LinearRegression()    # Régression linéaire
# model_LiR.fit(X_train, y_train)
# y_pred_lineaire = model_LiR.predict(X_test)
# print('# 2. Régression linéaire :')
# score = mean_squared_error(y_test, y_pred_lineaire, squared=False)
# # Affichez l'exactitude
# print(f"\n","Score :",score)


models = {
    "logistic": LogisticRegression,
    "tree": DecisionTreeClassifier,
    "forest": RandomForestClassifier,
    "svm": SVC
}

scores = [
    accuracy_score,
    precision_score,
    recall_score, 
    f1_score
]

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

results_tab = {}
for model_name, model in models.items():
    m = model()
    m.fit(X_train, y_train)
    y_pred = m.predict(X_test)
    results = []
    for score in scores:
        results.append(score(y_test, y_pred))
    results_tab[model_name] = results
    
pd.DataFrame.from_dict(results_tab, orient="index", columns=[str(n).split()[1] for n in scores]).round(2)
```
